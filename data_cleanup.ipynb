```python
import pandas as pd
```

```python
df = pd.read_csv('kayla_output.csv')
```

```python
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
df
```

```python
df_verified = df.groupby(['Date'])['count'].sum().reset_index()
df_verified = df_verified.rename(columns={"count": "count_all"})
df_verified
```

```python
df
```

```python
total = df_verified.merge(df, left_on=['Date'], right_on=['Date'])
total
```

```python
total.to_csv('counts_pc2.csv')
```
